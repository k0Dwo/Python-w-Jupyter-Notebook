```python
import pandas as pd
import requests
from io import StringIO

# Define the column names
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Get the response from the URL
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
response = requests.get(url)

# Read the CSV data from the response
iris_df = pd.read_csv(StringIO(response.text), names=column_names)

# Define the file path
file_path = r'C:\Users\bb\DataScienceLab\Lab 5 Visualisation\iris.csv'

# Save the dataframe to a CSV file
iris_df.to_csv(file_path, index=False)

# Print the first 10 rows of the dataframe
print(iris_df.head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Use the already loaded iris dataset
iris = iris_df
iris.head(30)
```

```python
sns.displot(iris['sepal_length'], kde=True)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Density")
plt.show()
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue ='species')
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
sns.barplot(x='species', y='sepal_length', data=iris, hue='species')
plt.title("Average Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length (cm)")
plt.show()
```

```python
sns.pairplot(iris, hue='species')
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

```python
# Exclude the 'species' column for correlation matrix calculation
corr_matrix = iris.drop(columns=['species']).corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=1)
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
# Customising scatter plot
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species', palette='dark')
plt.title("Customised Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True)
plt.show()
```
